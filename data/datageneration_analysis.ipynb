### Add Noise to the dataset

```python
import pandas as pd
import random
import string

# -----------------------------
# CONFIGURABLE PARAMETERS
# -----------------------------
OUTPUT_CSV = "support_tickets_noisy.csv"
TEXT_COLUMNS = ["title", "description","answer"]
DUPLICATE_RATIO = 0.1  # 10% of rows duplicated
NOISE_RATIO = 0.2       # 20% of rows will have noise added

# -----------------------------
# TEXT NOISE FUNCTIONS
# -----------------------------
def random_typo(text, n_typos=1):
    """Randomly introduce typos by swapping or deleting characters."""
    if not text or len(text) < 2:
        return text
    text = list(text)
    for _ in range(n_typos):
        idx = random.randint(0, len(text)-2)
        if random.random() < 0.5:
            # swap characters
            text[idx], text[idx+1] = text[idx+1], text[idx]
        else:
            # delete a character
            text.pop(idx)
    return ''.join(text)


def random_punctuation(text):
    """Randomly insert punctuation."""
    puncts = ['.', ',', '?', '!', '...', ':',"@","#"]
    words = text.split()
    for i in range(len(words)):
        if random.random() < 0.1:
            words[i] = words[i] + random.choice(puncts)
    return ' '.join(words)

def add_noise_to_text(text):
    """Combine multiple noise transformations."""
    if pd.isna(text):
        return text
    if random.random() < 0.5:
        text = random_typo(text, n_typos=random.randint(1, 2))
    if random.random() < 0.3:
        text = random_punctuation(text)
    return text

# -----------------------------
# MAIN PIPELINE
# -----------------------------
def add_noise_and_duplicates(df):
    df_noisy = df.copy()

    # Step 1: Add noise to a random subset
    noisy_indices = df_noisy.sample(frac=NOISE_RATIO, random_state=42).index
    for idx in noisy_indices:
        for col in TEXT_COLUMNS:
            df_noisy.loc[idx, col] = add_noise_to_text(df_noisy.loc[idx, col])

    # Step 2: Add duplicate rows
    n_dupes = int(len(df_noisy) * DUPLICATE_RATIO)
    duplicates = df_noisy.sample(n=n_dupes, random_state=42).copy()
    df_augmented = pd.concat([df_noisy, duplicates], ignore_index=True)

    return df_augmented

df_augmented = add_noise_and_duplicates(shuffled_df)
df_augmented.to_csv(OUTPUT_CSV, index=False)
print(f"✅ Augmented dataset saved to: {OUTPUT_CSV}")
print(f"Original rows: {len(shuffled_df)}, Final rows: {len(df_augmented)}")
```
